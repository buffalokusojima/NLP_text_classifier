```python
import sys
from gensim import matutils
from gensim import corpora
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
sys.path.append('../')
from util import util
from Dictionary import dictionary_class
```

```python
def vec2dense(vec, num_terms):
    return list(matutils.corpus2dense([vec], num_terms=num_terms).T[0])
```

```python
DATA_PATH = '../text'
OUTPUT_PATH = './dic'
```

```python
dic = dictionary_class.Dictionary(DATA_PATH, OUTPUT_PATH, no_below=100, no_above=0.5)
```

```python
word_list, dictionary = dic.make_medium_dictionary()
```

```python
print(dictionary)
```

```python
word_list = util.get_word_list_folder_label(DATA_PATH)
```

```python
print(word_list[0]['labels'])
```

```python
for words in word_list:
    for word in words['words']:
        word['text'] = vec2dense(dictionary.doc2bow(word['text']), len(dictionary))
```

```python
print(word_list[0]['words'][0])
```

```python
X_train = []
X_test = []
y_train = []
y_test = []

for words in word_list:
    x_tr, x_te, y_tr, y_te = train_test_split(words['words'], words['labels'], test_size=0.1, random_state=1, shuffle=True)
    X_train.extend(x_tr)
    X_test.extend(x_te)
    y_train.extend(y_tr)
    y_test.extend(y_te)

print('X_train',len(X_train))
print('X_test', len(X_test))
print('y_train', len(y_train))
print('y_test', len(y_test))
```

```python
train_file_name = [data['file'] for data in X_train]
X_train = [data['text'] for data in X_train]
test_file_name = [data['file'] for data in X_test]
X_test = [data['text'] for data in X_test]
```

```python
sc = StandardScaler()
sc.fit(X_train)
X_train_std = sc.transform(X_train)
X_test_std = sc.transform(X_test)
```

```python
clf = SVC(C = 100, kernel = 'rbf')
#clf.fit(X_train, y_train)
```

```python
clf.fit(X_train, y_train)
```

```python
score = clf.score(X_test_std, y_test)
print("{:.3g}".format(score))
```
